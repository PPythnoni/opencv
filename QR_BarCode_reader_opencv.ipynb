```python
import numpy as np
import cv2
from pyzbar.pyzbar import decode
```

```python
cam = cv2.VideoCapture("C:/Users/parti/Desktop/PY/opencv/qrs.mp4")
frames_counter = 1
while True:
    frames_counter = frames_counter + 1
    ret,frame = cam.read()
    if ret:
        width = 640
        height = 480
        dim = (width, height)
        frame = cv2.resize(frame, dim, interpolation=cv2.INTER_AREA)
        frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
        for barcode in decode(frame):
            print(barcode.data)
            txt_decoded = barcode.data.decode("utf-8")
            print(txt_decoded)
        
            pts = np.array([barcode.polygon],np.int32)
            pts = pts.reshape((-1,1,2))
            cv2.polylines(frame,[pts],True,(255,0,0),5)
            pts2 = barcode.rect
            cv2.putText(frame,txt_decoded,(pts2[0],pts2[1]),cv2.FONT_HERSHEY_SIMPLEX,0.8,(0,255,0),2)
            
        cv2.imshow("Result",frame)
        k = cv2.waitKey(1) & 0xFF
        if k == 27:
            break
            
            
    else:
        break
        
        
cam.release()
cv2.destroyAllWindows()
```
